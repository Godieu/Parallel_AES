```python
import matplotlib.pyplot as plt
import numpy as np

# Provided data
file_sizes = ["256B", "1KB", "4KB", "16KB", "64KB", "256KB", "1MB", "4MB", "16MB", "64MB", "256MB", "1GB"]
omp_times = [3.0208e-05, 0.000159273, 0.000627884, 0.00250437, 0.00367029, 0.0215089, 0.0610241, 0.237223, 0.946777, 3.75656, 15.1638, 60.6]
openssl_times = [4.779e-06, 3.3394e-05, 0.000103577, 0.00015253, 0.000774412, 0.00411506, 0.0150725, 0.0397667, 0.158226, 0.635996, 2.52452, 10.0733]

# Convert times to milliseconds
omp_times_ms = [time * 1000 for time in omp_times]
openssl_times_ms = [time * 1000 for time in openssl_times]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(file_sizes, omp_times_ms, marker='o', label='OMP')
plt.plot(file_sizes, openssl_times_ms, marker='o', label='OpenSSL')
plt.yscale('log')  # Use log scale for better visualization

# Adding labels and title
plt.xlabel('File Size')
plt.ylabel('Time (ms)')
plt.title('Encryption Time Comparison (OMP vs OpenSSL)')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
